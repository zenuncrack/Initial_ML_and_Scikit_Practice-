```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Basic Graph

```python
x =[1,2,3]
y=[2,4,6]
plt.plot(x,y, label='2x',linewidth='12')
plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
plt.figure(figsize=(5,3))

x =[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y, label='2x',color='red',marker='.',linestyle='--')
plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3,4])
#plt.yticks([2,3,4,5])




x2= np.arange(0,4,0.5)
plt.plot(x2,x2**2, color='green',linestyle='--')
```

##Barcharts


```python
labels = ['A','B','C']
values = [1,4,2]


bars = plt.bar(labels,values)
plt.figure(figsize=(6,4))

bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('*')

plt.show()
```
