```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.title('gas Prices Over Time')

plt.plot(gas['Year'],gas.USA,'b.-',label='Usa')
plt.plot(gas.Year,gas.Canada,'r.-', label='Canada')

plt.xticks(gas.Year[::3])
plt.legend()
plt.show()
```
