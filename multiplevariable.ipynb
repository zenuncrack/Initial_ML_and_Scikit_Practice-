```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# Define the data with column names and rows
columns = ['age', 'experience', 'risk']
data = [
    [10, 12, 2],
    [11, 3, 22],
    [31, 21, 2],
    [22, 0, 112],
    [10, 211, 221],
    [121, 2, 1]
]

# Create the DataFrame
df = pd.DataFrame(data, columns=columns)

# Save the DataFrame to a CSV file
df.to_csv('cars.csv', index=False)

# Create a Linear Regression model
reg = linear_model.LinearRegression()

# Fit the model
reg.fit(df[['age', 'experience']], df['risk'])

# Predictions
predictions = reg.predict(df[['age', 'experience']])

# Scatter plot of the original data
plt.scatter(df['age'], df['risk'], color='blue', label='Actual Risk')
plt.scatter(df['age'], predictions, color='red', label='Predicted Risk')

# Plot the linear regression line
plt.plot(df['age'], predictions, color='red')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Risk')
plt.title('Linear Regression: Age vs Risk')
plt.legend()

# Show the plot
plt.show()
```
